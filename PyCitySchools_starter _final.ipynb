```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
df = pd.DataFrame(school_data_complete)
num_of_schools=df['school_name'].nunique()
print(num_of_schools)
print(df.columns)
total_num_students = df['student_name'].count()
print(f"{total_num_students:,}")
total_budget =  pd.DataFrame(df.drop_duplicates(subset=['school_name','budget'], keep= 'last'))
TotalBudget = total_budget['budget'].sum() # total budget 
Avg_maths_score = df['maths_score'].mean()
AvgMathsScore= round(Avg_maths_score, 6)
Avg_reading_score = df['reading_score'].mean()
AvgReadingScore = round(Avg_reading_score, 6)
countmathspass = pd.DataFrame(df.loc[df['maths_score'] >=50,['school_name','maths_score']])
passing_MathsScore_percentage = (countmathspass['maths_score'].count()/total_num_students)*100
MathsScorePercent= round(passing_MathsScore_percentage, 6)
countreadingpass = pd.DataFrame(df.loc[df['reading_score'] >=50,['school_name','reading_score']])                                     
passing_ReadingScore_percentage = (countreadingpass['reading_score'].count()/total_num_students)*100
ReadingScorePercent = round(passing_ReadingScore_percentage, 6)
overall_pass = (len(df[(df['reading_score']>= 50)&(df['maths_score']>=50)])/total_num_students)*100
OverallPassPercent = round(overall_pass, 6)
LGAS_df = pd.DataFrame([
    {"Total Schools":num_of_schools,
     "Total Students": f"{total_num_students:,}",
     "Total Budget": f"${TotalBudget:,.2f}",
     "Average Maths Score":AvgMathsScore,
     "Average Reading Score":AvgReadingScore,"% Passing Maths":MathsScorePercent,"% Passing Reading":ReadingScorePercent,
    "% Over passing":OverallPassPercent}
])
LGAS_df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python

school_type = school_data.set_index("school_name")["type"]
school_student_num = df.groupby("school_name").size()
total_budget = df.groupby("school_name")["budget"].mean()
perstudentbudget = total_budget/school_student_num
avgmathsscore =  df.groupby("school_name")["maths_score"].mean()
avgreadingscore =  df.groupby("school_name")["reading_score"].mean()
passing_maths_df = df.loc[df['maths_score'] >=50,['school_name','maths_score']]
percentpassingmaths = passing_maths_df.groupby("school_name").size()/school_student_num *100
passing_reading_df = df.loc[df['reading_score'] >=50,['school_name','reading_score']]
percentpassingreading = passing_reading_df.groupby("school_name").size()/school_student_num *100
overallpassing_df= df.loc[(df['reading_score']>= 50)&(df['maths_score']>=50)]
overall_passing = overallpassing_df.groupby("school_name").size()/school_student_num *100
school_summary = pd.DataFrame({"type": school_type,"studentsperschool":school_student_num,
                               "totalbudget": total_budget,"perstudentbudget":perstudentbudget,
                               "avgmathsscore":avgmathsscore, "avgreadingscore": avgreadingscore,
                               "%passing maths score":percentpassingmaths, "% passing reading":percentpassingreading,
                               "overall_passing %":overall_passing, 
                              })

school_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
SummaryBySize = school_summary.sort_values("studentsperschool", ascending = False)
AvgOverallPassing= (SummaryBySize["avgmathsscore"] + SummaryBySize["avgreadingscore"])/2
AvgOverallPassingPercent= AverageOverallPassing *100
SummaryBySize["AvgOverallPassingPercent"] = round(AvgOverallPassingPercent, 2)
bins = [59.9,69.9, 79.9, 89.9]
group_schools = [ "C", "B", "A"]
SummaryBySize["Ranking"] = pd.cut(SummaryBySize["overall_passing %"], bins, labels=group_schools, include_lowest =True)
SummaryByRanking= SummaryBySize.sort_values("Ranking", ascending = False)
SummaryByRanking
```
